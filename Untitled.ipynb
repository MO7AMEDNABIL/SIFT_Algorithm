```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('sphinx2.jpg')  #read image
plt.imshow(img1) #show image
```

```python
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY) #convert the image to gray scale
plt.imshow(img1_gray, cmap='gray') #show the image and set the color map to gray color
```

```python
#pip install opencv-contrib-python 
```

```python
sift = cv2.xfeatures2d.SIFT_create() #using & define sift algorthim, there is way used to download it in anaconda by using conda install -c menpo opencv
kp_1, descriptors_1 = sift.detectAndCompute(img1,None) #use to detect and compute the keypoints on image
```

```python
img1_draw = cv2.drawKeypoints(img1_gray,kp_1,img1) #draw the key points on the image
plt.imshow(img1_draw) 
```

```python
img2 = cv2.imread('sphinx1.jpg')   #read image
plt.imshow(img2)  #show image
```

```python
img2_gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY) #convert the image to gray scale

#show the image and set the color map to gray color
plt.imshow(img2_gray, cmap='gray')
```

```python
figure, ax = plt.subplots(1, 2, figsize=(16, 8)) # make one row two coulmns to set on it my image

ax[0].imshow(img1_gray, cmap='gray') #image set img1 in first coulmn
ax[1].imshow(img2_gray, cmap='gray') #image set img2 in second coulmn
```

```python
sift = cv2.xfeatures2d.SIFT_create() #using & define sift algorthim

keypoints_1, descriptors_1 = sift.detectAndCompute(img1_gray,None) #apply SIFT on image1 on gray scale
keypoints_2, descriptors_2 = sift.detectAndCompute(img2_gray,None) #apply SIFT on image2 on gray scale
```

```python
flann = cv2.FlannBasedMatcher() #fun used to show the matches between two images

# Perform the matching
matches = flann.knnMatch(descriptors_1, descriptors_2, k=2)

# Apply ratio test
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)

img3 = cv2.drawMatches(img1_draw, keypoints_1, img2_gray, keypoints_2, good_matches[:50], img2, flags=2)
plt.imshow(img3),plt.show()
```
